```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load raw dataset
df = pd.read_csv("../data/raw/student_data.csv")
df.head()
```

```python
# Remove target variables from input features 
targets = ["G2", "G3"]
X = df.drop(columns=targets)
y = df[targets]
```

```python
# Convert categorical variables into numerical format using one-hot encoding
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
# Normalize all input features using Min-Max scaling, this ensures all features are on a comparable scale
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X_encoded),
    columns=X_encoded.columns
)
```

```python
# Recombine processed input features with target variables
df_cleaned = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
df_cleaned.shape
```

```python
# Save the processed dataset
df_cleaned.to_csv("../data/processed/student_data_cleaned.csv", index=False)
```
